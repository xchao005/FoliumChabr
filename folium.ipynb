## 1.3 Prostorová data

Prostorová (geografická) data jsou data, která obsahují informaci o poloze na Zemi.

Každý objekt má:
- geografickou polohu (souřadnice)
- atributy (např. název, typ, hodnota)

### Typy prostorových dat

#### Vektorová data
Reprezentují objekty pomocí geometrie:

- **Body** – např. města, zastávky  
- **Linie** – např. silnice, řeky  
- **Polygony** – např. státy, kraje  

#### Rastrová data
Reprezentují data jako mřížku pixelů:

- satelitní snímky  
- letecké snímky  
- výškové modely  

### Proč jsou důležitá?

Prostorová data umožňují analyzovat:
- kde se něco nachází  
- jaké jsou vztahy mezi objekty  
- jak se data mění v prostoru

## 2.1 Historie a pozadí

Folium vzniklo jako nástroj, který propojuje Python s JavaScriptovou knihovnou Leaflet.js.

### Leaflet.js
Leaflet je open-source JavaScriptová knihovna pro tvorbu interaktivních map ve webovém prostředí.

- vznik kolem roku 2011  
- velmi lehká a rychlá  
- široce používaná ve webových aplikacích  

Folium funguje jako „wrapper“:
- uživatel píše Python  
- ale na pozadí se generuje JavaScript (Leaflet)  

To znamená, že:
- není potřeba znát JavaScript  
- ale stále využíváme jeho možnosti

### FINÁLNÍ PŘÍKLAD bez mých dat: Prodejna v Praze s 3mi pobočkami a chce po kliknutí na prodejnu sdělit základní informace

```python
import folium

m_branches = folium.Map(location=[50.0755, 14.4378], zoom_start=12)
```

```python
branches = [
    {
        "name": "Pobočka Praha 1",
        "lat": 50.0870,
        "lon": 14.4200,
        "address": "Na Příkopě 12, Praha 1",
        "hours": "Po–Pá 9:00–18:00"
    },
    {
        "name": "Pobočka Praha 4",
        "lat": 50.0500,
        "lon": 14.4390,
        "address": "Nuselská 45, Praha 4",
        "hours": "Po–Pá 8:00–17:00"
    },
    {
        "name": "Pobočka Praha 7",
        "lat": 50.1050,
        "lon": 14.4450,
        "address": "Dukelských hrdinů 21, Praha 7",
        "hours": "Po–So 10:00–20:00"
    }
]
```

```python
for branch in branches:
    popup_text = f"""
    <b>{branch['name']}</b><br>
    Adresa: {branch['address']}<br>
    Otevírací doba: {branch['hours']}
    """
    
    folium.Marker(
        location=[branch["lat"], branch["lon"]],
        popup=folium.Popup(popup_text, max_width=300),
        tooltip=branch["name"],
        icon=folium.Icon(
            color="red",           # barva
            icon="shopping-cart",  # ikona
            prefix="fa"            # font awesome
        )
    ).add_to(m_branches)
```

```python
m_branches

m_branches.save("pobocky_praha.html")
```

### FINÁLE: Hustota obyvatelstva ČR z mých dat

```python
import geopandas as gpd
import folium
```

```python
# 1) načtení shapefile
gdf = gpd.read_file("kraje.shp")
```

```python
# 2) převod do WGS84 pro webovou mapu
gdf = gdf.to_crs(epsg=4326)
```

```python
# 3) přepočet plochy z m² na km²
gdf["AREA_km2"] = gdf["AREA"] / 1_000_000
```

```python
# 4) výpočet hustoty obyvatelstva
gdf["hustota_obyv"] = gdf["OB91"] / gdf["AREA_km2"]
```

```python
# 5) vytvoření mapy
m_hustota = folium.Map(location=[49.8, 15.5], zoom_start=7)
```

```python
# 6) choropleth s jemnějším rozdělením horní části škály
folium.Choropleth(
    geo_data=gdf,
    data=gdf,
    columns=["NAZEV", "hustota_obyv"],
    key_on="feature.properties.NAZEV",
    fill_color="YlOrRd",
    fill_opacity=0.7,
    line_opacity=0.3,
    legend_name="Hustota obyvatelstva (obyv./km²)",
    bins=[0, 80, 100, 120, 140, 160, 200, 400, 800, 1600, 2500]
).add_to(m_hustota)
```

```python
# 7) obrys + tooltip s atributy
folium.GeoJson(
    gdf,
    style_function=lambda feature: {
        "fillColor": "transparent",
        "color": "black",
        "weight": 1
    },
    tooltip=folium.GeoJsonTooltip(
        fields=["NAZEV", "OB91", "AREA_km2", "hustota_obyv"],
        aliases=["Kraj:", "Obyvatel:", "Plocha (km²):", "Hustota:"],
        localize=True
    )
).add_to(m_hustota)
```

```python
# 8) zobrazení mapy
m_hustota

# 9) export
m_hustota.save("hustota_obyvatelstva.html")
```
